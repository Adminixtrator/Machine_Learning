```python
file = open('Autism-Adult-Data.arff', 'r')
raw_data = file.read()
file.close()
raw_data.split('\n')
```

```python
cleaned_data = []

import re

regex = re.compile('@{1}\W*')

for all in raw_data.split('\n'):
    if all != '' and all != ' ' and regex.findall(all) == []:
        cleaned_data.append(all)
        
cleaned_data[0]
```

```python
new_data = open('data.txt', 'w')

for data in cleaned_data:
    content = open('data.txt', 'r').read()
    new_data.write(data+'\n')
        
new_data.close()
```
